#### Removal punctuations and numbers

```python
#1 Remove Punctuations
#2 Remove numbers
# I will also convert all words to lower case

def remove(docus):
    #convert words to lower case
    lower_c = map(str.lower, docus)
    
    #1. remove punctuations
    remove_punc = list(map(lambda x: " ".join(re.findall('\\b\\w\\w+\\b',x)),lower_c))
    
    #2. remove numbers
    return list(map(lambda x:re.sub('\\b[0-9]+\\b', '', x), remove_punc))

docs = remove(docs)
print(docs)
```

### Lemmatization

##### In the block below I will do lemmatization and in the process of doing this I will also remove stop words and removes words that appear in only one document and words that appear in over 90% of the document. 

```python
from nltk import word_tokenize
from nltk.stem import WordNetLemmatizer

#3 lemmatization 
class LemmaTokenizer(object):
    def __init__(self):
        self.wnl = WordNetLemmatizer()
    def __call__(self, doc):
        return [self.wnl.lemmatize(t, 'v') for t in word_tokenize(doc)]
#3 lemmatization
#4 remove stop words
#5 remove words that only appear in only one document
#6 Remove words that appear in over 90% of the documents
vectorizer_norm = TfidfVectorizer(tokenizer=LemmaTokenizer(), lowercase=True, stop_words = 'english', 
                            max_df=.90, min_df=2,norm=None)
X_norm = vectorizer_norm.fit_transform(docs)
df_tf_norm = pd.DataFrame(X_norm.todense(), columns=vectorizer_norm.get_feature_names())

print(X_norm)
print(df_tf_norm)
```

# TF-IDF

##### TF-IDF stands for term frequency (TF) and inverse document frequency (IDF). TF-IDF is an information retrieval technique that incorporates the term frequency the ineverse document frequency to produce a weight or score. In some sense, the score represents the occurance of each word; a lower score means the word/term is recurrent, the higher a score the word/term is scarce.   The term frequency (TF) is the number of times the word appears in a document. The inverse document frequency (IDF) measures the signficance of the word/term in a entire corpus. 

##### Lets break this down a littlle further. If the word "data" appears 100 times in 1,000 word document the TF would be the 10/1000 = 0.10. This means the term frequency is 10% in that 1000 word document.
